```python
test_accuracy_file = ["u10_00_0", "u20_00_0", "u40_00_0"]
```

```python
def read_data_from_csv(result_dir, file_type, file_list):
    li = []
    for file in file_list:
        file_path = result_dir + file + file_type
        df = pd.read_csv(file_path, index_col= None, header=0)
        li.append(df)

    data = pd.concat(li, axis=0, ignore_index=True)
    return data 
```

```python
dynamic_accuracy_data = read_data_from_csv(dynamic_result_dir, "_solving_stats.csv", test_accuracy_file)
```

```python
mlbin_accuracy_data = read_data_from_csv(mlbin_result_dir, "_solving_stats.csv", test_accuracy_file)
```
